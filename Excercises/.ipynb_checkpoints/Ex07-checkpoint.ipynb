```python
P = np.array([[0.0,0.0],[0.1,0.8], [0.2,-0.8], [0.3 ,0.2],[0.4,0.3],[0.5,-0.9],[0.6,0.4],[0.7,-0.4],[0.8 ,0.1],[0.9, -0.6],[1.0 ,1.0]])
for it in range(1001):
    if(it % 250 == 0):
            print(f"iteration : {it}")
            x = P[:,0]
            y = P[:,1]
            plt.plot(x,y,label=f"It: {it}")
            
    for i in range(1,len(P)-1):
        grad = 2 * (2 * P[i] - P[i - 1] - P[i + 1])
        P[i] =  P[i] -0.01 * grad

plt.xlabel= "x"
plt.ylabel = "y"
plt.legend()
plt.show()
plt.savefig("SmoothingOverTime")
```
